**Observations**

- Series can form a dictionary when we type cast dict

- Series can also easily convert into dataframe

```python
df = pd.DataFrame()
df['datatype'] = visa_df.dtypes
df.reset_index(inplace = True)
df.rename({'index' : 'Columns', 'datatype' : 'DataTypes'}, axis = 1, inplace = True)
df
```

### Extract Categorical columns and numerical columns in separate list

```python
cat_cols,num_cols = [],[]
for i,j in visa_df.dtypes.to_dict().items():
    if j == 'object':
        cat_cols.append(i)
    else:
        num_cols.append(i)
```

```python
print('Categorical Columns :', cat_cols)
print('Numerical Columns :', num_cols)
```

```python
cat = [i for i,j in visa_df.dtypes.to_dict().items() if j == 'object']
numr = [i for i,j in visa_df.dtypes.to_dict().items() if j != 'object']
cat,numr
```
